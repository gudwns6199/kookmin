```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/인공지능부트캠프/week1/BitCoin.csv')
data
```

```python
idx = []
for i in range(len(data)):
    a = data['Date'][i]
    a = list(a.split('-'))
    a[0] = int(a[0])
    a[1] = int(a[1])
    if a[0] <= 2017 or a[0] >= 2016:
        if a[0] == 2017:
            if a[1] <= 6:
                idx.append(i)
        elif a[0] == 2016:
            if a[1] >= 6:
                idx.append(i)
```

```python
data = data.iloc[idx,:]
date_BIT = data['Date']
date_BIT = list(date_BIT)
```

```python
average = []
for i in range(len(idx)-4):
    sum = data['Open'][idx[i]] + data['Open'][idx[i+1]] + data['Open'][idx[i+2]] + data['Open'][idx[i+3]] + data['Open'][idx[i+4]]
    average.append(sum/5)
```

```python
idx = [i * 320 for i in range(6)]
```

```python
average = list(reversed(average))
date = data['Date'][:-4]
date = list(reversed(list(date)))
```

```python
x = []
for i in range(6):
    x.append(date[i*70])
```

```python
plt.plot(date,average,color = '#f2a900')
plt.xticks(x)
```

```python
data_ETH = pd.read_csv('/content/drive/MyDrive/인공지능부트캠프/week1/ETH_day.csv')
data_ETH
```

```python
idx_ETH = []
for i in range(len(data_ETH)):
    if data_ETH['Date'][i] in date_BIT:
        idx_ETH.append(i)
```

```python
average_ETH = []
average_ETH_10 = []
for i in range(len(idx_ETH)-4):
    sum = data_ETH['Open'][idx_ETH[i]] + data_ETH['Open'][idx_ETH[i+1]] + data_ETH['Open'][idx_ETH[i+2]] + data_ETH['Open'][idx_ETH[i+3]] + data_ETH['Open'][idx_ETH[i+4]]
    average_ETH.append(sum/5)
    average_ETH_10.append(sum*10/5)
```

```python
average_ETH = list(reversed(average_ETH))
average_ETH_10 = list(reversed(average_ETH_10))
```

```python
plt.figure(figsize=(10,20))

plt.plot(date,average,color = '#f2a900',label = 'BIT')
plt.plot(date,average_ETH,color = '#3c3c3d',label = 'ETH')
plt.plot(date,average_ETH_10,color = 'r', label = 'ETH x 10')

plt.legend()
plt.xticks(x)
plt.show()
```
