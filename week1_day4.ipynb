```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/인공지능부트캠프/week1/netflix_titles.csv')
south_korea = data[data['country']=="South Korea"]
south_korea
print('South Korea Netflex : {}'.format(len(south_korea)))
```

```python
num = np.ones(len(data))
num = {'num' : num}
num = pd.DataFrame(num)
data = pd.concat([num, data], axis =1)

country = pd.DataFrame(data['country'])
country = country.fillna(value = '1,1')

idx = []
for i in range(len(country)):
    a = country['country'].iloc[i]
    a = a.split(',')
    if len(a) == 1:
        idx.append(i)

data = data.iloc[idx,:]

data = data['num'].groupby(by = data['country'])
data = data.sum()

dict_data = dict(data)
country = list(dict_data.keys())
value = list(dict_data.values())

for idx, num in enumerate(data):
    if num == data.max():
        print('country : {}, num : {}'.format(country[idx], value[idx]))
        break
```
