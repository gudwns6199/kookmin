```python
x = np.arange(-10,10,0.01)

plt.plot(x,x**2)

# x,y 축에 설명 첨가하기
plt.xlabel('x value')
plt.ylabel('y = f(x) value')

# 주목할 부분만 출력하기위해 범위 설정 ; 주목할 부분의 최소 최대 value를 인자로 받음
plt.axis([-5, 5, 0, 25])

# 각 축의 눈금을 표시할 value들을 설정
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 27, 3)])

plt.show()
```

```python
x = np.arange(-10,10,0.01)

plt.plot(x,x**2)

# x,y 축에 설명 첨가
plt.xlabel('x value')
plt.ylabel('y = f(x) value')

# 주목할 부분만 출력하기위해 범위 설정 ; 주목할 부분의 최소 최대 value를 인자로 받음
plt.axis([-5, 5, 0, 25])

# 각 축의 눈금을 표시할 value들을 설정
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 27, 3)])

# 그래프에 title을 설정
plt.title('y = f(x) = x^2 (graph)')

plt.show()
```

```python
x = np.arange(-10,10,0.01)

# 그래프의 주석 첨가(legend)
plt.plot(x,x**2, label = 'trend')
plt.legend()

# x,y 축에 설명 첨가
plt.xlabel('x value')
plt.ylabel('y = f(x) value')

# 주목할 부분만 출력하기위해 범위 설정 ; 주목할 부분의 최소 최대 value를 인자로 받음
plt.axis([-5, 5, 0, 25])

# 각 축의 눈금을 표시할 value들을 설정
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 27, 3)])

# 그래프에 title을 설정
plt.title('y = f(x) = x^2 (graph)')

plt.show()
```

## 3. Matplotlib Case Study

- 꺽은선 그래프 ( plot )
- plt.plot()

```python
x = np.arange(20)
y = np.random.randint(0,20,20)

plt.plot(x,y)

plt.axis([0,20,0, 21])
plt.yticks([0,5,10,15,20])

plt.show()
```

- 산점도(산포도) 그래프 ( scatter )
- plt.scatter()

```python
plt.scatter(x,y)
plt.show()
```

- 박스 그림 ( Box )
- 수치형 데이터(min, Q1, Q2, Q3, max)
- plt.box()

```python
plt.boxplot(y)

plt.title("Box plot of Y")

plt.show()
```

```python
plt.boxplot( (x,y) )

plt.title("Box plot of X , Y")

plt.show()
```

- 막대 그래프 ( bar )
- 해당 값과 해당 값의 수치를 표시
- plt.bar()

```python
plt.bar(x,y)

plt.xticks(x)

plt.show()
```

- hisrogram
- 도수 그래프를 직사각형의 막대 형태로 나타냄
- **계급**추가 ( 0, 1, 2 -> 0 ~ 2 )
- plt.hist()

```python
plt.hist(y,bins = np.arange(0,22,2))

plt.xticks(np.arange(0,22,2))

plt.show()
```

- 원형 그래프 (Pie chart)
- 전체 data 중 차지하는 비율을 확인할 수 있음
- plt.pie()

```python
z = [100, 300, 200, 400]
plt.pie(z, labels = z)

plt.show()
```

### import seaborn as sns

```python
import seaborn as sns
```

- 커널 밀도 그림
- plt.hist() 와 유사함


- sns.kdeplot()

```python
x = np.arange(0, 22, 2)
y = np.random.randint(0,20,20)

plt.hist(y, bins = x)
plt.show()
```

```python
sns.kdeplot(y, shade = True) # shade 는 그래프 밑 부분을 음영처리함

plt.show()
```

- 카운트 그림
- 범주형 column의 빈도수를 시각화


- sns.countplot()

```python
vote_df = pd.DataFrame({'name': ['Andy', 'Bob', 'Cat'], 'vote':[True, True, False]})

vote_df
```

```python
vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x = [False, True], height = vote_count['name'])
```

```python
sns.countplot(x = vote_df['vote'])

plt.show()
```

- 캣 그림
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여줌


- sns.catplot()

```python
covid = pd.read_csv('C:/Users/gudwn/OneDrive/바탕 화면/country_wise_latest.csv')

covid.head(5)
```

```python
s = sns.catplot(x= 'WHO Region', y= 'Confirmed', data=covid)
s.fig.set_size_inches(10, 6)
plt.show()
```

- 스트립 그림( strip plot)
- plt.scatter() 와 유사한 그래프를 출력


- sns.stripplot()

```python
sns.stripplot(x= 'WHO Region',y= 'Recovered',data=covid)

plt.show()
```

```python
# stripplot과 동일하지만 뭉쳐있는 data들을 옆으로 분산시켜 시각적으로 읽기 쉬운 형태로 그래프를 출력함

sns.swarmplot(x= 'WHO Region',y= 'Recovered',data=covid)

plt.show()
```

- 히트맵 ( heatmap )
- 데이터의 행렬을 색상으로 표현해주는 그래프


- sns.heatmap()

```python
covid.corr()
```
