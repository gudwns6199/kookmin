## Mission:

### 1) covid data에서 100case 대비 사망률이 가장 높은 국가를 찾으시오

```python
covid_country = covid['Deaths / 100 Cases'].groupby(by = covid['Country/Region'])

max_value = max(covid_country.max())
covid_country = covid_country.max()

for idx, value in enumerate(covid_country):
    if value == max_value:
        break
covid.loc[idx,'Country/Region']
```

### 2) covid data에서 신규 확진자가 없는 나라 중 WHO Region이 'Europe'인 나라를 모두 출력하시오

```python
zero_cases = covid[ covid['New cases'] == 0 ]
Europe_zeros_cases = zero_cases[ zero_cases['WHO Region'] == 'Europe' ]

print(Europe_zeros_cases['Country/Region'])
```
